NumPy. CRUD operations over data set with 0/1/2-dimensional arrays

```python
import numpy as np

mainArray = np.arange(11,42,10) # Create 1-dimensional array(vector)
mainArray[3] = 51 # Update last number
mainArray = np.delete(mainArray, -1) # Delete last number
mainArray = np.append(mainArray, 41) # Append 0-dimensional array(scalar) to vector
mainArray = np.vstack([mainArray, np.round(np.random.random(4),1)]) # Create 2-dimensional array(matrix) from 2x 1-dimensional
print(mainArray)
print(mainArray.ndim)
```

Pandas. Creating dataframes based on dict and array, render head, describe, locate/index locate dataframes' methods

```python
import pandas as pd

dataDict = {"Cat": 5, "Dog": 4, "Raccoon": 3, "Monkey": 3}
dataArray = [["Cat", 5], ["Dog", 4], ["Raccoon", 3], ["Monkey", 3]]
mainFrame = pd.DataFrame(dataArray, columns=["Pet", "Rate"])
viaSeriesFrame = pd.Series(dataDict).to_frame("Rate")
print(f"#{mainFrame}\n ${mainFrame.head(1)}\n ${mainFrame.describe()}\n ")
print(f"#{viaSeriesFrame}\n ${viaSeriesFrame.iloc[0]}\n ${viaSeriesFrame.loc['Dog']}")
```

Matplotlib

```python
import matplotlib.pyplot as plt
import csv

x = np.array([1, 2, 3, 4, 5])
y = np.round(np.random.random(5),1)
plt.plot(x, y, label='Випадкова функція')
plt.title('Функція випадкових значень')
plt.xlabel('X')
plt.ylabel('Y')

y_pred = np.round(np.random.random(5),1)

mae = np.mean(np.abs(y - y_pred))
mse = np.mean((y - y_pred)**2)
print(f'MAE: {mae}')
print(f'MSE: {mse}')

file = open('./results.csv', 'w', newline='')
csvFile = csv.writer(file)
csvFile.writerow([f'X: {x}, Y: {y}, Predicted Y: {y_pred}'])
csvFile.writerow([f'MAE: {mae}, MSE: {mse}'])
file.close()
```
