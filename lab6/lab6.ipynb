Tensors Create (C) and Read (R) operations in different dimentions

```python
import tensorflow as tf
import matplotlib.pyplot as plt

scalar = tf.constant(10)  
vector = tf.constant([-1, 4, 7, 31, 6])  
matrix = tf.constant([[11, 21, 31], [41, 11, 21], [11, 21, 31]])

print("Scalar:", scalar.numpy())
print("Vector:", vector.numpy())
print("Matrix:\n", matrix.numpy())


print("Zero index in vector", vector[0].numpy())
print("Element [0, 1] of matrix:", matrix[0, 1].numpy())
```

Update (U) & Delete (D) operations over tensors

```python

vector = tf.tensor_scatter_nd_update(vector, [[3], [2]], [16, 18])
matrix = tf.tensor_scatter_nd_update(matrix, [[0, 1], [1, 1]], [41, 11])
print("Updated vector:", vector.numpy())
print("Updated matrix:\n", matrix.numpy())

vector = tf.concat([vector[:0], vector[1:]], axis=0)
matrix = tf.concat([matrix[:1], matrix[2:]], axis=0)
print("Vector without second element:", vector.numpy())
print("Matrix without second row:\n", matrix.numpy())
```

Linear Regression random dataset generated by TensorFlow & visulization

```python
X_regression = tf.random.uniform([100, 1], minval=0, maxval=10, dtype=tf.float32)
y_regression = 3 * X_regression + 2 + tf.random.normal([100, 1], mean=0.0, stddev=1.0, dtype=tf.float32)

plt.scatter(X_regression, y_regression, color='blue', label='Фактичні дані')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Дані з набору для регресії')
plt.legend()
plt.show()
```

Classification random dataset generated by TensorFlow & visulization

```python
class_zero = tf.random.normal([50, 2], mean=2.0, stddev=1.0)  
class_one = tf.random.normal([50, 2], mean=6.0, stddev=1.0) 

X_classification = tf.concat([class_one, class_zero], axis=0)
y_classification = tf.constant([0] * 50 + [1] * 50, dtype=tf.int32)

plt.scatter(X_classification[:, 0], y_classification, color='blue', label='Фактичні дані')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Дані з набору для класифікації')
plt.legend()
plt.show()
```
