Regression datasets. Kaggle dataset pull, dataset creation with DatasetGenerator, random dataset creation. Datasets preparing

```python
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

## Kaggle dataset

possum_dataset_url = kagglehub.dataset_download("abrambeyer/openintro-possum")
possum_datafile = pd.read_csv(os.path.join(possum_dataset_url, "possum.csv"))
X_kaggle = possum_datafile[['totlngth']]
y_kaggle = possum_datafile['hdlngth']
X_kaggle = X_kaggle.fillna(X_kaggle.median())
y_kaggle = y_kaggle.fillna(y_kaggle.median())

## DatasetGenerator dataset

X, y = make_regression(n_samples=100, n_features=10, noise=10)
data = pd.DataFrame(X, columns=[f'Feature_{i}' for i in range(1, 11)])
data['Target'] = y

X = data.drop(columns=[f'Feature_{i}' for i in range(1, 11)])

## Random dataset

dataManual = {
    'X': np.arange(0, 100, 1),
    'y': np.random.random(100)
}

df = pd.DataFrame(dataManual)

X_numpy = df[['X']]
y_numpy = df['y']
```

Linear Regression Kaggle dataset visualization

```python

model_kaggle = LinearRegression()
model_kaggle.fit(X_kaggle, y_kaggle)
y_pred_k = model_kaggle.predict(X_kaggle)

mse = mean_squared_error(y_kaggle, y_pred_k)
mae = mean_absolute_error(y_kaggle, y_pred_k)
r2 = r2_score(y_kaggle, y_pred_k)

print(f"Середньоквадратична похибка (MSE): {mse}")
print(f"Середня абсолютна похибка (MAE): {mae}")
print(f"Коефіцієнт детермінації (R^2): {r2}")

plt.scatter(X_kaggle, y_kaggle, color='blue', label='Фактичні дані')
plt.plot(X_kaggle, y_pred_k, color='red', linewidth=2, label='Лінійна регресія')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Лінійна регресія для завантаженого набору')
plt.legend()
plt.show()
```

Linear Regression for dataset created with DatasetGenerator

```python
model_generator = LinearRegression()
model_generator.fit(X, y)
y_pred = model_generator.predict(X)

mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"Середньоквадратична похибка (MSE): {mse}")
print(f"Середня абсолютна похибка (MAE): {mae}")
print(f"Коефіцієнт детермінації (R^2): {r2}")

plt.scatter(X, y, color='blue', label='Фактичні дані')
plt.plot(X, y_pred, color='red', linewidth=2, label='Лінійна регресія')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Лінійна регресія для завантаженого набору')
plt.legend()
plt.show()
```

Linear Regression for random dataset

```python
model_random = LinearRegression()
model_random.fit(X_numpy, y_numpy)
y_pred_np = model_random.predict(X_numpy)

mse = mean_squared_error(y_numpy, y_pred_np)
mae = mean_absolute_error(y_numpy, y_pred_np)
r2 = r2_score(y_numpy, y_pred_np)

print(f"Середньоквадратична похибка (MSE): {mse}")
print(f"Середня абсолютна похибка (MAE): {mae}")
print(f"Коефіцієнт детермінації (R^2): {r2}")

plt.scatter(X_numpy, y_numpy, color='blue', label='Фактичні дані')
plt.plot(X_numpy, y_pred_np, color='red', linewidth=2, label='Лінійна регресія')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Лінійна регресія для завантаженого набору')
plt.legend()
plt.show()
```
