```python
import matplotlib.pyplot as plt
import tensorflow as tf
import pandas as pd
import numpy as np
import kagglehub
import os
```

Linear Regression for Possum regression with TensorFlow and Neural Networks

```python

possum_dataset_url = kagglehub.dataset_download("abrambeyer/openintro-possum")
possum_datafile = pd.read_csv(os.path.join(possum_dataset_url, "possum.csv"))

X_kaggle = tf.constant(possum_datafile[['totlngth']])
y_kaggle = tf.constant(possum_datafile['hdlngth'])

model = tf.keras.Sequential([
  tf.keras.layers.Dense(1)
])

model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae', 'mse']
)


md = model.fit(X_kaggle, y_kaggle, epochs=2000)
y_pred_k = model.predict(X_kaggle)

plt.figure(figsize=(10, 7))
plt.scatter(X_kaggle, y_kaggle, color='blue', label='Фактичні дані')
plt.scatter(X_kaggle, y_pred_k, color='red', label='Лінійна регресія')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Лінійна регресія для завантаженого набору')
plt.legend()
plt.show()


model.evaluate(X_kaggle, y_kaggle)
pd.DataFrame(md.history).plot()
plt.ylabel("loss")
plt.xlabel("epochs")

```

Classification for Cat&Dogs Kaggle dataset with TensorFlow and Neural Networks

```python
dataFile = pd.read_csv('cat_dog.csv')

X_kaggle = tf.constant(dataFile.drop('image', axis=1))
y_kaggle = tf.constant(dataFile['labels'])

model_classification = tf.keras.Sequential([
  tf.keras.layers.Dense(1)
])

model_classification.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer=tf.keras.optimizers.SGD(),
    metrics=['accuracy']
)

x_min, x_max = X_kaggle[:, 0].numpy().min() - 0.1, X_kaggle[:, 0].numpy().max() + 0.1
y_min, y_max = X_kaggle[:, 0].numpy().min() - 0.1, X_kaggle[:, 0].numpy().max() + 0.1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 250),
                       np.linspace(y_min, y_max, 100))
  
x_in = np.c_[xx.ravel(), yy.ravel()] # stack 2D arrays together: https://numpy.org/devdocs/reference/generated/numpy.c_.html

model_classification.fit(x_in, y_kaggle, epochs=50)


y_pred = model_classification.predict(x_in)

if len(y_pred[0]) > 1:
  print("doing multiclass classification...")
  y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)
else:
  print("doing binary classifcation...")
  y_pred = np.round(y_pred).reshape(xx.shape)
  
plt.contourf(xx, yy, y_pred, cmap=plt.cm.RdYlBu, alpha=0.7)
plt.scatter(X_kaggle[:, 0], y=y_kaggle, s=40, cmap=plt.cm.RdYlBu)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
```
