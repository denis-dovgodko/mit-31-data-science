Task 1

Токенізація - розбиття тексту на так звані лексеми

Лемматизація та стемінг - методи нормалізації тексту для подальшої обробки мови моделями.
Лемматизація аналізує проводить морфологічний аналіз, наприклад, замінює дієприкметники на дієслова, інше
Стемінг прибирає суфікси слів, намагаючись зробити з них щось схоже на іменники, хоч і не завжди вдало

Векторизація - процес перетворення тексту-строки на послідовність чисел у вигляді вектору.
Bag of Words працює зі словником і співставляє слова відповідним числам
TF-IDF додатково до попереднього ще й розставляє вагові коефіцієнти до слів відносно частоти появи в тексті, Word2Vec шукає контекст вжитих слів

Класифікація тексту займається визначенням загальної категорії, до якої можна віднести текст. Наприклад, за жанром, або за емоційним забарвленням, типом літератури, інше

Розпізнавання сутностей (Named Entity Recognition, NER) працює з метою визначення в мові конкретних діячів, істот, місць, які важливі в тексті, для їх подальшої класифікації. Сутністю може бути країна, місто, валюта, число, дата, інше географічне місце, особа, інше

Наївний байєсівський класифікатор вважає всі слова в тексті незалежними

LSTM це рекурентна модель NN, зосереджена на контексті висловів і відношень слів між собою

Transformers це модель NN, що зосереджує увагу на головних(ключових) моментах тексту

GPT це модель NN яка генерує високоякісні тексти
